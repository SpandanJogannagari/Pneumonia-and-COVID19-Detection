```python
df['Modal_Accuracy'] = (df['tp']/(df['fp']+df['tp']))* 100
```

```python
df['Covid_Accuracy'] = (((df['fp']+df['tp'])-df['covid_num']-df['wrong_covid']+df['correct_covid'])/((df['fp']+df['tp'])-df['covid_num']+df['correct_covid']+df['not_detected_covid']))*100
df['covid_Sensitivity']=(df['correct_covid']/df['covid_num'])*100
df['covid_Specificity']=(((df['fp']+df['tp'])-df['covid_num']-df['wrong_covid'])/((df['fp']+df['tp'])-df['covid_num']))*100
```

```python
df['Pnemonia_Sensitivity']=(df['correct_pneu']/df['pneu_num'])*100
df['Pnemonia_Accuracy'] = (((df['fp']+df['tp'])-df['pneu_num']-df['wrong_pneu']+df['correct_pneu'])/((df['fp']+df['tp'])-df['pneu_num']+df['correct_pneu']+df['not_detected_pneu']))*100
df['Pnemoia_Specificity']=(((df['fp']+df['tp'])-df['pneu_num']-df['wrong_pneu'])/((df['fp']+df['tp'])-df['pneu_num']))*100
```

```python
df['Normal_Sensitivity']=(df['correct_normal']/df['normal_num'])*100
df['Normal_Specificity']=(((df['fp']+df['tp'])-df['normal_num']-df['wrong_normal'])/((df['fp']+df['tp'])-df['normal_num']))*100
df['Normal_accuracy']=(((df['fp']+df['tp'])-df['normal_num']-df['wrong_normal']+df['correct_normal'])/((df['fp']+df['tp'])-df['normal_num']+df['correct_normal']+df['not_detected_normal']))*100
```
